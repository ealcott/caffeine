```python
import pandas as pd
import numpy as np
```

```python
starbucks = pd.read_csv("starbucks_drinkMenu_expanded.csv")
```

```python
starbucks.head()
```

```python
#this column has different kinds of information, especially if I will be examining certain nutrients by beverage.
#I will split this column into different columns, Size & Prepared Beverages
starbucks["Beverage_prep"].unique()
```

```python
#creating "Size" column
size_list = ["Short", "Grande", "Tall", "Venti"]
for size in size_list:
    starbucks.loc[starbucks["Beverage_prep"].str.contains(size), "Size"] = size
starbucks.head(15)
```

```python
#accounting for espresso sizes and null values (where size is dependent on placement in table)
for i in starbucks[starbucks["Size"].isnull()].index:
    starbucks.loc[i, "Size"] = starbucks.loc[i - 1, "Size"]
starbucks.loc[starbucks["Beverage_prep"].str.contains("Solo"), "Size"] = "Solo (Espresso)"
starbucks.loc[starbucks["Beverage_prep"].str.contains("Doppio"), "Size"] = "Doppio (Espresso)"
```

```python
starbucks[starbucks["Beverage"] == "Espresso"]
```

```python
#use regular expressions to find preparations for "Prepared Beverages" column
import re
new_names = []
for row_index in range(len(starbucks["Beverage_prep"])):
    if re.findall("Nonfat Milk", starbucks["Beverage_prep"].iloc[row_index]):
        new_name = starbucks["Beverage"].iloc[row_index] + ", Nonfat Milk"
        new_names.append(new_name)
    elif re.findall("2% Milk", starbucks["Beverage_prep"].iloc[row_index]):
        new_name = starbucks["Beverage"].iloc[row_index] + ", 2% Milk"
        new_names.append(new_name)
    elif re.findall("Soymilk", starbucks["Beverage_prep"].iloc[row_index]):
        new_name = starbucks["Beverage"].iloc[row_index] + ", Soymilk"
        new_names.append(new_name)
    elif re.findall("Whole Milk", starbucks["Beverage_prep"].iloc[row_index]):
        new_name = starbucks["Beverage"].iloc[row_index] + ", Whole Milk"
        new_names.append(new_name)
    else:
        new_names.append(starbucks["Beverage"].iloc[row_index])
print(new_names)
```

```python
#creating "Prepped Beverages" column
starbucks["Prepped Beverages"] = new_names
starbucks.head(15)
```

```python
def replace_percents(col_name):
    try:
        starbucks[col_name] = starbucks[col_name].str.replace("%", "").astype(float)/100
    except Exception:
        pass
```

```python
for col in starbucks.columns[-7:-3]:
    replace_percents(col)
```

```python
starbucks.head()
```

```python
#clean up various formatting errors discovered when uploading to SQL database (below)
starbucks["Vitamin A (% DV) "].unique()
```

```python
starbucks["Caffeine (mg)"].unique()
starbucks["Caffeine (mg)"] = starbucks["Caffeine (mg)"].replace('varies', np.nan)
starbucks["Caffeine (mg)"] = starbucks["Caffeine (mg)"].replace('Varies', np.nan)
starbucks = starbucks.dropna()
starbucks["Caffeine (mg)"].unique()
```

```python
starbucks.columns
starbucks[" Total Fat (g)"].unique()
starbucks[" Total Fat (g)"] = starbucks[" Total Fat (g)"].replace('3 2', '3.2')
starbucks[" Total Fat (g)"].unique()
```

```python
import mysql.connector as mysql
```

```python
db = mysql.connect(host = 'localhost',
                              user = 'root',
                              passwd = '',
                              database = 'starbucks_db'
                                     )
```

```python
cur = db.cursor(prepared=True)
```

```python
cur.execute("DROP TABLE IF EXISTS drink_data")
cur.execute('''
        CREATE TABLE IF NOT EXISTS drink_data (
            Beverage_category TEXT,
            Beverage TEXT,
            Beverage_prep TEXT,
            Calories INT,
            Total_fat_g FLOAT,
            Trans_fat_g FLOAT,
            Saturated_fat_g FLOAT,
            Sodium_mg FLOAT,
            Total_carbohydrates_g INT,
            Cholesterol_mg INT,
            Dietary_fiber_g INT,
            Sugars_g INT,
            Protein_g INT, 
            Vitamin_A_DV FLOAT,
            Vitamin_C_DV FLOAT, 
            Calcium_DV FLOAT,
            Iron_DV FLOAT,
            Caffeine_mg INT,
            Size TEXT,
            Prepped_beverages TEXT
            
        )
        
''')
print("inserting rows")

cur.executemany('''
    INSERT INTO drink_data 
        (Beverage_category, Beverage, Beverage_prep, Calories, Total_fat_g, Trans_fat_g,
        Saturated_fat_g, Sodium_mg, Total_carbohydrates_g, Cholesterol_mg, Dietary_fiber_g,
        Sugars_g, Protein_g, Vitamin_A_DV, Vitamin_C_DV, Calcium_DV, Iron_DV, Caffeine_mg, Size, Prepped_beverages)
        
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    
    
''', 
               starbucks.values.tolist())

db.commit()

sql = "SELECT * FROM drink_data"
cur.execute(sql)

result = cur.fetchall()
for i in result:
    print(i)
```
